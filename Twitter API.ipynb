```python
import os
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk import word_tokenize          
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk import wordpunct_tokenize

vect = CountVectorizer(tokenizer = tokenize, stop_words = en_stop, ngram_range= (1,3), lowercase= True,max_features = 150)

dtm = vect.fit_transform(listdf)

print(dtm.shape)

df_dtm = pd.DataFrame(dtm.toarray(), columns = vect.get_feature_names())
print(df_dtm.head())
```

```python
pd.set_option('display.max_columns', 500)
df_dtm.head()
```

```python
#creating an extra column
df_dtm['Person'] = 0
```

```python
#assigning the y columns
df_dtm['Person'][0:500] = 0
df_dtm['Person'][500:1000] = 1
df_dtm['Person'][1000:1500] = 2
df_dtm['Person'][1500:2000] = 3 
df_dtm['Person'][2000:2500] = 4
```

```python
#df_dtm.isnull().sum()
df_dtm = df_dtm.fillna(0)
```

```python
#retrieving required column indices for X and Y
y_index=df_dtm.columns.get_loc("Person")
```

```python
y_index
```

```python
import numpy as np
dtm=df_dtm.as_matrix()

X=np.delete(dtm,y_index,axis=1)

Y=dtm[:,y_index]
```

```python
y = pd.factorize(Y)[0]  #Encode input values as an enumerated type or categorical variable
len(y)

# Frequency distribution
from scipy import stats
stats.itemfreq(y) # displays how many times 0 is repeated and 1 is repeated
```

```python
from sklearn.cross_validation import train_test_split
import numpy

x_train ,x_test,y_train,y_test = train_test_split(X,y,test_size=0.4, random_state=42) 
```

```python
# Basic code for setting up a model
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
import numpy as np
```

```python
#One-hot encoding: Needed if the categorical variable is stored as numeric
from keras.utils import to_categorical

#y_cat = to_categorical(y_train)
y_cat = pd.get_dummies(y_train)

# Save the number of columns in predictors: n_cols
n_cols = x_train.shape[1]
print(n_cols)
# Set up the model: model
model = Sequential()

# Add the first layer
model.add(Dense(80, activation='relu', input_shape=(n_cols,)))
# Add the second layer
model.add(Dense(52, activation='relu'))
# Add the output layer

model.add(Dense(5, activation = 'softmax'))

#sgd = optimizers.SGD(lr=0.1)
#print(sgd)
model.compile(optimizer='adam',loss='categorical_crossentropy')

# Confirming that the model compiled correctly 
#print("Loss function: " + model.loss)

# Fitting the model
#model.fit(X, y)

#adding this
model.fit(x_train,y_cat,epochs = 10)


# Compile the model


# Fit the model
#model.fit(Xepa, yepa_cat)
```

```python
# Calculate predictions: predictions
predictions = model.predict(x_test)
print(predictions)

# # Calculate predicted probability of positive: predicted_prob_true
predicted_prob_true = predictions[:,1]

# # print predicted_prob_true
print(predicted_prob_true)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

predictions = model.predict_classes(x_test)
print('\n')
print(pd.Series(predictions).value_counts(), '\n')
print(pd.Series(y_test).value_counts(), '\n')

print('Accuracy on test data:', accuracy_score(y_test, predictions), '\n')
print('Confusion matrix: ', '\n', confusion_matrix(y_test, predictions), sep = '')
```
